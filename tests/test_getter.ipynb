```python
import requests

r = requests.get('https://api.github.com/events')
# criamos um objeto-Response (response é uma resposta à uma requisição http) 'r' que contém as informações do site do github 

#print(type(r))
#print(r)
r.text
```

```python
 r.encoding
```

```python
r.content
```

```python
# Binary response

from PIL import Image
from io import BytesIO

i = Image.open(BytesIO(r.content))
```

```python
# JSON response

import requests
r = requests.get('https://api.github.com/events')
r.json()
```

```python
# Raw response

import requests
r = requests.get('https://api.github.com/events', stream=True)
r.raw
r.raw.read(10)
```

```python
with open("arquivo", 'wb') as fd:
    for chunk in r.iter_content(chunk_size=128):
        fd.write(chunk)
```
